```python
# Function to assign labels based on scores
def assign_label(scores,x):
    if max(scores) >= 0.60:
        return x['labels'][0]
    else:
        return 'None'
```

```python
candidate_labels=['Gender','Muslim','Tamil']
candidate_labels1=['Positive content','Negative content','Neutral Content']
candidate_labels2=['biased','reasonable']
```

```python
labels=[]
sentiment=[]
classify=[]
max_length = 512
i=0
for row in df1:
  if len(row) > max_length:
    row = row[:max_length]
  x=classifier(row,candidate_labels)
  y=analyzer(row)
  z=classifier(row,candidate_labels2)
  label=assign_label(x['scores'],x)
  labels.append(label)
  sentiment.append(y[0]['label'])
  classify.append(z['labels'][0])
  print('The total text labelled are: '+str(i+1))
  i=i+1

```

```python
i=0
for index in range(len(labels)):
  if labels[index]!='None':
    i=i+1
print("the total number of data points labelled are "+str(i))
```

```python
df2=df[df_start_index:df_end_index]
```

```python
df2['Labels']=labels
df2['Sentiment']=sentiment
df2['classify']=classify
#df2.head()
```

```python
df3 = df2[['Text', 'Labels','Sentiment','classify']]
```

```python
for i in range(len(df3)):
  if df3['classify'][i]=='reasonable':
    df3['classify'][i]='Non Discriminative'
  else:
    df3['classify'][i]='Discriminative'
```
